```python
from bs4 import BeautifulSoup
import requests
import re
import urllib.parse
import os
from pathlib import Path
```

```python
page_url = "https://wiki.leagueoflegends.com/en-us/Category:A_Perfect_Life"

reg = re.findall(".*Category:(.*)", page_url)[0]
book_name = urllib.parse.unquote(reg)

os.mkdir(book_name)

response = requests.get(page_url)

soup = BeautifulSoup(response.content)
```

```python
for el in soup.find_all("span", attrs={"typeof":"mw:File"}):
    link = list(el.children)[0]["href"]
    reg = re.findall(".*File:(.*)", link)[0]
    img_url = f"https://wiki.leagueoflegends.com/en-us/images/{reg}"

    response = requests.get(img_url)
    with open(Path(book_name) / Path(reg), "wb") as f:
        f.write(response.content)
```
