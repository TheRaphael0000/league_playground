```python
from collections import defaultdict

txt = open("matches.txt").read().split("\n")

team_position = defaultdict(list)

for i in range(len(txt)):
    position = i % 12
    team = txt[i]
    pos = int((position) / 2) + 1
    team_position[team].append(pos)


team_position = dict(team_position)
print(team_position)
mean_ = {k: sum(v)/len(v) for k, v in team_position.items()}
print(mean_)
team_by_mean = [t for t, v in sorted(list(mean_.items()), key=lambda p: p[-1])]
print(team_by_mean)
```

```python
import matplotlib.pyplot as plt

lookup_time = {
    1: "16:45",
    2: "17:30",
    3: "18:15",
    4: "19:00",
    5: "19:45",
    6: "20:30",
}

lookup_name = {
    'LR': "Los Ratones",
    'FNC': "Fnatic",
    'SHFT': "Shifters",
    'GX': "GIANTX",
    'TH': "Team Heretics",
    'NAVI': "Natus Vincere",
    'VIT': "Team Vitality",
    'SK': "SK Gaming",
    'KCB': "Karmine Corp Blue",
    'KC': "Karmine Corp",
    'MKOI': "Movistar KOI",
    'G2': "G2 Esports",
}

plt.figure(figsize=(16, 12))
for i, team in enumerate(team_by_mean):
    poses = team_position[team]
    plt.subplot(3, 4, i+1)

    plt.title(f"{lookup_name[team]} ({team})")
    plt.hist(poses, bins=[1, 2, 3, 4, 5, 6, 7],
             align='left', rwidth=0.6, label=["Matches"])
    plt.ylim(0, 7)
    plt.xticks(list(lookup_time.keys()), labels=lookup_time.values())
    plt.xlabel("Scheduled start of match [UTC+1]")
    plt.ylabel("# Match")
    plt.vlines(mean_[team], 0, 7, colors="red", label="Avg. start time")
    plt.legend()
    plt.grid(axis='y')

plt.tight_layout()
plt.savefig("output/lec_2026_versus_start_of_match.png")
plt.show()
```
